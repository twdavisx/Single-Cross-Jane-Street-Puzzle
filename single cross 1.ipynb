```python
import math
import random
import matplotlib.pyplot as plt
```

```python
def nLinesCrossed(i,k,theta, l):
    x = i + l * math.cos(theta)
    y = k + l * math.sin(theta)
    
    return abs(math.floor(x)) + abs(math.floor(y))
```

```python
def MonteCarloSim(iters, l):
    sum = 0
    for t in range(iters):
        # random coordinates in bottom left quadrant.
        i = random.random() * 0.5
        k = random.random() * 0.5
        theta = random.random() * 2 * math.pi
        if nLinesCrossed(i,k,theta, l) == 1:
            sum+=1
    return sum/iters
```

```python
l = range(0,2000)
results = []

for i in l:
    results.append(MonteCarloSim(1000,i/1000))
```

```python
plt.figure(figsize=(10, 5))
plt.bar(l, results, width=0.9, color='skyblue')
plt.xlabel("length")
plt.ylabel("P(crosses 1 line)")
plt.title("Monte Carlo Simulation")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
print(MonteCarloSim(10000000,1))
```
